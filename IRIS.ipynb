## Selecting X and Y (Target variable)

```python
y=df['Species']
x=df.drop('Species',axis=1)
```

```python
x.head()
```

```python
y.head()
```

## Label Encoding for target Variable

```python
label_encoder = LabelEncoder()
label_encoder.fit(y)
labeled_y = label_encoder.transform(y)
```

```python
labeled_y
```

```python
for i in range(0,3):
    print(f"{i} Represent {label_encoder.inverse_transform([i])}")
```

## Spliting Data in Test and Train 

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Implementing Logistic Regression for classification 

```python
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)
y_pred = log_reg.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

## Implementing Naive Bayes for classification 

```python
model = GaussianNB()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('Accuracy:', accuracy)
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True)
```

## Implementing Random Forest for classification 

```python
clf = RandomForestClassifier(n_estimators=100, random_state=1)
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print("Accuracy:", clf.score(X_test, y_test))
cm = confusion_matrix(y_test, y_pred)
```

```python
sns.heatmap(cm, annot=True)
plt.title("Confusion Matrix")
plt.xlabel("Predicted Labels")
plt.ylabel("True Labels")
plt.show()
```
